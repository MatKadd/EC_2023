We are given three columns of integers, with a row for each node. The first two columns contain x
and y coordinates of the node positions in a plane. The third column contains node costs. The goal is
to select exactly 50% of the nodes (if the number of nodes is o,dd we round the number of nodes to
be selected up) and form a Hamiltonian cycle (closed path) through this set of nodes such that the
sum of the total length of the path plus the total cost of the selected nodes is minimized.
The distances between nodes are calculated as Euclidean distances rounded mathematically to
integer values. The distance matrix should be calculated just after reading an instance and then only
the distance matrix (no nodes coordinates) should be accessed by optimization methods to allow
instances defined only by distance matrices.

# Experminets


When calculating distances, we incorporated node costs into them, so basically distance(i, j) = euclidean_distance(i, j) + node_cost(j).

```python
def analyze_instance(instance):
    distances = calculate_distances(instance)

    nn_scores, nn_best = get_scores(get_nearest_neighbor_solution, distances)
    gc_scores, gc_best = get_scores(get_greedy_cycle_solution, distances)
    random_scores, random_best = get_scores(get_random_solution, distances)

    print(f"Statistics for: {instance.name}")
    visualize_stats(random_scores, "Random solution")
    visualize_solution(instance, random_best)

    visualize_stats(nn_scores, "Nearest neighbor")
    visualize_solution(instance, nn_best)

    visualize_stats(gc_scores, "Greedy cycle")
    visualize_solution(instance, gc_best)
```

Run experiments for each instance and show their results.

```python
for instance in instances:
    analyze_instance(instance)
```

# Conclusions

From the results, we can clearly see that random solution is by far the worst one. Greedy cycle is a bit better than nearest neighbor in terms of quality of solution, but it has much worse performance due to its computational complexity.
